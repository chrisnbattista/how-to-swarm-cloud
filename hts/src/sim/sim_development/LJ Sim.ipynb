```python
import lj_test
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig = None
ax = None
try:
    get_ipython().magic("matplotlib notebook")
    sns.set_theme()
    ##fig, ax = plt.subplots(1, 1)
    fig = plt.figure(figsize=(12,12))
    plt.ion()
except Exception as e:
    lj_test.setup_plots()
    print("oops")
    print(e)
```

```python
def lj_notebook_data_viz_w_fig(world, long_world_history, i, fig, ax, label):
    if i % 100 == 0:
        fig.clf()
        p2 = sns.scatterplot(
            x=world[:,1],
            y=world[:,2],
            s=1000,
            color='k'
        )
        fig.suptitle(label)
        fig.canvas.draw()
        plt.show()
        plt.pause(0.00001)

lj_notebook_data_viz = lambda world, long_world_history, i, label: lj_notebook_data_viz_w_fig(world, long_world_history, i, fig, ax, label)
```

```python
lj_test.run_sim(data_viz=lj_notebook_data_viz)
```
